*   보스턴 집값 관련 데이터라고 생각해볼 경우 
*   13개 column의 Input Layer, 5개 column의 Hidden Layer, 1개 column의 output layer라고 볼 수 있다.



## Hidden Layer 추가 코드

### Hidden layer 없는 구조

```python
# <Hidden layer가 없는 구조>

X = tf.keras.layer.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

### Hidden layer 있는 구조

```python
# <Hidden layer가 있는 구조>
X = tf.keras.layer.Input(shape=[13])
H = tf.keras.layers.Dense(5, activation='swish')(X)  
# 히든레이어 활성화 함수에 'swish 활성화 함수' 사용 # 5: 히든레이어가 5개의 노드임을 뜻함 
H = tf.keras.layers.Dense(3, activation='swish')(H)
H = tf.keras.layers.Dense(3, activation='swish')(H)
# 이렇게 여러개의 히든레이어(3개의 노드)를 추가하여 학습을 더욱 정교하게 시킬 수 있다
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```
