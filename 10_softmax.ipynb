*   softmax: 비율로 예측하기위해 사용함

softmax.png



*   (setosa, virginica, versicolor) = (0.7, 0.3, 0) 의 경우 setosa일 확률 70% virginica일 확률 30%로 볼 수 있다
*   이처럼 함수값이 0~1사이의 값이 나오게 하고 싶을 경우 각 함수를 softmax로 감싸주면 된다.



softmax_1.png



*   입력을 그대로 출력으로 만드는 identity(y=x)함수


```python
# 1. 분류모델의 구조
X = tf.keras.layer.Input(shape=[4])
Y = tf.keras.Dense(3, activation='softmax')(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy')


# 2. 회귀모델의 구조
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```



*   위: 분류모델 , 아래 : 회귀모델
*   분류모델에서는 loss로 나오는 값이 crossentropy를 통해 계산된 차이값을 이용해 나온 것



```python
# 분류모델의 구조
X = tf.keras.layer.Input(shape=[4])
Y = tf.keras.Dense(3, activation='softmax')(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy'
              metrics = 'accuracy')
```



*   metrics = 'accuracy'를 추가함에 따라 결과에서 정확도도 같이 확인할 수 있다.
*   accuracy 값 1 = 모두 맞음, accuracy 값 0 = 모두 틀림

## !!) 분류문제는 회귀문제와 달리 확률문제가 등장하므로 어려울 수 있다.
