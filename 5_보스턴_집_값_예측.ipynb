1.   과거의 데이터를 준비한다
2.   모델의 구조를 만든다
3.   데이터로 모델을 학습(FIT)한다
4.   모델을 이용한다



## 1) 과거의 데이터를 준비한다.

```python
보스턴 = pd.read_csv('boston.csv')
독립 = 보스턴[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
종속 = 보스턴['medv']
print(독립.shape, 종속.shape)
```

## 2) 모델의 구조를 만든다

```python
X = tf.keras.layers.Input(shape=[13]) # 13 -> 13개의 독립변수
Y = tf.keras.layers.Dense(1)(X) # 1 -> 1개의 종속변수
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

*   퍼셉트론의 기본 모형



퍼셉트론_1.png

*   위 퍼셉트론은 하나의 뉴런(neuron)으로 볼 수 있다.



퍼셉트론_2.png

*   12개의 독립변수와 2개의 종속변수의 경우 위와 같이 그려볼 수 있다.

## 3) 데이터로 모델을 학습(FIT)





```python
model.fit(독립, 종속, epochs=1000)
```

## 4) 모델을 이용한다

```python
print("predictions: ", model.predict(독립[0:5]))
```
