# 1. 변수 타입 확인 및 변경

```python
import pandas as pd
```

```python
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head()
```



* 품종 data가 virginica, setosa 같이 명목형 데이터가 
* setosa=0, virginica=1, versicolor=2 와 같은 숫자형 데이터로 바뀐상태 
* but still 품종의 종류를 표시하는 **범주형** 데이터인 상태
* 따라서 실수형 데이터로 바꾸는 원핫인코딩 필요



## 1) 원핫인코딩

```python
# 원핫인코딩

인코딩 = pd.get_dummies(아이리스)
인코딩.head()
```



*   but 원핫인코딩이 되지 않는다 왜냐하면 실제 data값은 실수로 적혀있기 때문이다



## 2) 변수의 타입확인

```python
print(아이리스.dtypes)
```



*   float: 실수형, int: 정수형
*   object 또는 category의 경우에만 원핫인코딩가능



## 3) 변수를 범주형으로 변경

```python
# 품종 타입을 범주형으로 바꾸어 준다.

아이리스['품종'] = 아이리스['품종'].astype('category')

print(아이리스.dtypes)
```

```python
# 원핫인코딩

인코딩 = pd.get_dummies(아이리스)
인코딩.head()
```



*   원핫인코딩이 성공적으로 마무리



# 2. NA 값의 처리

## 1) NA 갯수 체크

```python
# NA 값을 체크

아이리스.isna().sum()
```

```python
아이리스.tail()
```


*   149번째 행에 NaN(NA)값 확인
*   위 데이터를 모델에 그대로 넣으면 error 발생하므로 NA값을 바꾸어 줘야함



## 2) NA 값 채우기

```python
# NA값에 꽃잎폭의 평균값을 넣어준다
mean = 아이리스['꽃잎폭'].mean()
아이리스['꽃잎폭'] = 아이리스['꽃잎폭'].fillna(mean) # NA값을 mean으로 채워라
아이리스.tail()
```

*   149번째 행에 NaN(NA)값을 '꽃잎폭'의 평균값으로 바꿈
