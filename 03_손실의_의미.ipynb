# 3. 손실의 의미

>model을 학습(FIT)시킨 후 결과에서 손실(loss)의 의미

## 학습(FIT)의 결과에서 손실(loss)의 의미loss의 의미.png

 

1.   독립변수와 종속변수로 모델을 만듦
2.   독립변수를 모델에 넣으면 예측 값이 나옴
3.   모델 평가를 위해 종속변수, 예측 값 비교
4.   (예측-결과)^2 값을 만들고 값들의 평균을 구한다. (평균=loss)
5.   예측이 정답을 모두 맞추고 있으면 평균이 0이 나온다.
